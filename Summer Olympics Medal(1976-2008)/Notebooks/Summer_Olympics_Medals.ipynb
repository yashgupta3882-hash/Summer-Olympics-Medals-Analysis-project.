```
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Loading the dataset
df = pd.read_csv('/content/Summer-Olympic-medals-1976-to-2008.csv', encoding='latin1')
df
```

```
#Data Cleaning
df.isnull().sum()
```

```
df_cleaned = df.dropna()
df_cleaned
```

```
df_cleaned.info()
```

```
#EDA
# Total Medal Count by Country
medals_by_country = df_cleaned.groupby('Country')['Medal'].count().sort_values(ascending=False)
medals_by_country
```

```
plt.figure(figsize=(10,6))
medals_by_country.head(10).plot(kind='bar',color='gold')
plt.title('Top 10 Countries with most Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.show()
```

```
# Medals Won Over the Years
medals_over_years = df_cleaned.groupby('Year')['Medal'].count()
medals_over_years
```

```
plt.figure(figsize=(12,10))
plt.plot(medals_over_years.index, medals_over_years.values, marker='o', linestyle = '-', color = 'blue')
plt.title('Medals Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

```
# Gender Distribution in Events
gender_distribution = df_cleaned['Gender'].value_counts()
gender_distribution
```

```
plt.figure(figsize=(8,6))
gender_distribution.plot(kind='pie', colors=['blue','red'],explode = [0.05,0])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```
#Top Atheletes with most medals
athelete_medal_count = df_cleaned.groupby('Athlete')['Medal'].count().sort_values(ascending=False)
athelete_medal_count
```

```
plt.figure(figsize=(10,8))
athelete_medal_count.head(10).plot(kind='bar',color='green')
plt.title('Top 10 Athletes with most Medals')
plt.xlabel('Athlete')
plt.ylabel('Total Medals')
plt.show()
```
